#### Working With File Paths
When working with files in Python, handling file paths correctly is crucial to ensure your code works across different operating systems and environments. Python provides several modules and functions for working with file paths effectively.

```python
#### Using the os module
import os
cwd=os.getcwd() ## getcwd() is a function from the os module that returns the current working directory.
# It shows the path of the folder where the Python script is currently running.
print(f"Current working directory is {cwd}")
```

```python
## create a new directory
new_directory="package"
os.mkdir(new_directory)
print(f"Directory '{new_directory}' create")
```

```python
## Listing Files And Directories
items=os.listdir('.') #This function from the os module returns a list of all entries (files and folders) in the specified directory.
print(items)
```

```python
### Joining Paths

dir_name="folder"
file_name="file.txt"
full_path=os.path.join(dir_name,file_name) #this joins the folder and file name into a proper path depending on your operating system:
print(full_path)
```

```python

dir_name="folder"
file_name="file.txt"
full_path=os.path.join(os.getcwd(),dir_name,file_name) #joins the current directory, the folder name, and the file name into a full absolute path.
print(full_path)
```

```python
path='example1.txt'
if os.path.exists(path): # checks if the file or folder exists at the given path.
    print(f"The path '{path}' exists")
else:
    print(f"The path '{path}' does not exists")
```

```python
#Checking if a Path is a File or Directory
import os

path = 'example.txt'
if os.path.isfile(path): #if the path exists and is a file
    print(f"The path '{path}' is a file.")
elif os.path.isdir(path): #checks if it is a folder.
    print(f"The path '{path}' is a directory.")
else:
    print(f"The path '{path}' is neither a file nor a directory.")
```

```python
## Getting the absolute path
relative_path='example.txt'
absolute_path=os.path.abspath(relative_path) #Converts the relative path into an absolute path. Absolute path is the full path from the root of your file system to the file.
print(absolute_path)
```
