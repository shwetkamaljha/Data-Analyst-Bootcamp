#### File Operation- Read And Write Files

File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, both text and binary. This lesson will cover the basics of file handling, including reading and writing text files and binary files.

```python
### Read a Whole File

with open('example.txt','r') as file:    #  It opens the file safely for reading, provides access to its contents, and ensures it is automatically closed once the operation is complete.
    content=file.read()  
    print(content)
```

```python
## Read a file line by line
with open('example.txt','r') as file:   # 'r' stands for read mode
    for line in file:
        print(line.strip()) ## sstrip() removes the newline character
```

```python
## Writing a file(Overwriting)

with open('example.txt','w') as file: # 'w' stands for write mode and this line opens the file for writing, erases existing content if any, creates it if it doesn't exist, and automatically closes it when done.
    file.write('Hello World!\n')
    file.write('This is a new line.\n')
```

```python
## Write a file(wwithout Overwriting)
with open('example.txt','a') as file: # Opens the file for appending, keeps existing content, creates it if it doesn't exist, and automatically closes it when done.
    file.write("Append operation taking place!\n")
```

```python
### Writing a list of lines to a file
lines=['First line \n','Second line \n','Third line\n']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
### Binary Files

# Writing to a binary file
data = b'\x00\x01\x02\x03\x04'  # # Defines a bytes object containing raw binary data.
with open('example.bin', 'wb') as file: # Opens the binary file for writing, erases existing content if any, creates it if it doesn't exist, and automatically closes it when done.
    file.write(data)
```

```python
# Reading a binary file
with open('example.bin', 'rb') as file: # Opens the binary file for reading, allowing access to raw bytes, and automatically closes it when done.
    content = file.read()
    print(content)
```

```python
### Read the content froma  source text file and write to a destination text file
# Copying a text file
with open('example.txt', 'r') as source_file:
    content = source_file.read()

with open('destination.txt', 'w') as destination_file:
    destination_file.write(content)

# Reads the entire content from 'example.txt' and writes it into 'destination.txt'; if 'destination.txt' doesn't exist, it is created first, then the content is copied.
```

```python
#Read a text file and count the number of lines, words, and characters.
# Counting lines, words, and characters in a text file
def count_text_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines) #split() :->  Breaks the text into a list of words based on whitespace.
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count

file_path = 'example.txt'
lines, words, characters = count_text_file(file_path)
print(f'Lines: {lines}, Words: {words}, Characters: {characters}')
```

The w+ mode in Python is used to open a file for both reading and writing. If the file does not exist, it will be created. If the file exists, its content is truncated (i.e., the file is overwritten).

```python
### Writing and then reading a file

with open('example.txt','w+') as file:  # Opens the file for both writing and reading; creates the file if it doesn't exist and erases existing content, and automatically closes it when done.    file.write("Hello world\n")
    file.write("Hello World!\n")
    file.write("This is a new line \n")

    ## Move the file cursor to the beginning
    file.seek(0) ## Moves the file pointer to the beginning of the file, setting where read/write operations start.

    ## Read the content of the file
    content=file.read()
    print(content)
```
