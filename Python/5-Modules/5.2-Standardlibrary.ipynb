* Array Module

The array module in Python is a built-in module used to create arrays that store elements of the same data type.

Unlike lists, which can store mixed data types, arrays are typed, meaning all elements must be of one specific type (like only integers or only floats).

```python
import array
arr=array.array('i',[1,2,3,4]) # 'i' stands for integer, and the list [1,2,3,4] is the initial content of the array. array.array() means importing array function from array module and creating an array object with type 'i' and initial values [1,2,3,4].
print(arr)
print(type(arr))
print(arr[0])
print(arr[3])
```

```python
from array  import array
import math
arr=array('f',[1,2.5,3.3,4]) # 'f' stands for float, and the list [1,2.5,3.3,4] is the initial content of the array. from array import array means importing only the array function from the array module and creating an array object with type 'f' and initial values [1,2.5,3.3,4].
print(arr)
print(type(arr))
print(arr[0])
print(arr[3])
lr=list(map(math.sqrt , arr)) # map() is a built-in function in Python that applies a given function to each item of an iterable (like a list) and returns an iterator. In this case, math.sqrt is the function being applied to each element of the array arr. The result is a new list lr that contains the square root of each element in arr.
print(lr)
```

** math module 

The math module is a built-in Python module that provides mathematical functions and constants for numerical calculations.

You don’t need to install it. Just write:

* import math

```python
import math
print(math.sqrt(16)) # sqrt() is a function from the math module that calculates the square root of a number.
print(math.pi)  # pi() is a function from the math module that returns the value of pi (approximately 3.14159).
print(math.cos(0))
print(math.factorial(5)) # factorial() is a function from the math module that calculates the factorial of a number. The factorial of a non-negative integer n is the product of all positive integers less than or equal to n. For example, factorial(5) would return 120, since 5! = 5 * 4 * 3 * 2 * 1 = 120.
print(math.pow(2,3)) # pow() is a function from the math module that calculates the power of a number. It takes two arguments: the base and the exponent. For example, math.pow(2, 3) would return 8.0, since 2 raised to the power of 3 is 8 (2 * 2 * 2 = 8). Note that math.pow() returns a float, even if the result is an integer.
```

* random module

The random module is a built-in Python module used to generate random numbers and random selections.

```python
## random 

import random
print(random.randint(1,10)) # randint() is a function from the random module that generates a random integer between the specified range. In this case, it will return a random integer between 1 and 10 (inclusive).
print(random.choice(['apple','banana','cherry'])) # The random module is a built-in Python module used to generate random numbers and random selections. The choice() function is a method from the random module that takes a sequence (like a list) as an argument and returns a randomly selected element from that sequence. In this case, it will randomly select and return one of the strings 'apple', 'banana', or 'cherry' from the provided list.
```

* OS Module 

The os module in Python is a built-in module that allows you to interact with the operating system.

It helps you work with:

Files and folders

Directories (create, delete, rename)

Environment variables

System-related tasks

```python
### File And Directory Access

import os #Operating System module
print(os.getcwd()) ## getcwd() is a function from the os module that returns the current working directory.
# It shows the path of the folder where the Python script is currently running.
```

```python
os.mkdir('test_dir') # mkdir() is a function from the os module that creates a new directory (folder).
# In this case, it will create a folder named 'test_dir' in the current working directory.
```

* shutil Module

The shutil module in Python is used for high-level file and directory operations.

Common Uses of shutil:

shutil.copy() → Copies a file

shutil.move() → Moves a file or folder

shutil.rmtree() → Deletes a directory and all its contents

shutil.copytree() → Copies an entire directory

```python
## High level operations on files and collection of files
import shutil
shutil.copyfile('source.txt','destination.txt') # copyfile() is a function from the shutil module that copies the contents of one file to another file.

 # In this case, it will copy all data from 'source.txt' and automatically create (or overwrite) the 'destination.txt' file with the same content.
```

* JSON 

JSON (JavaScript Object Notation) is a lightweight data format used to store and exchange data.
It is text-based and language-independent, which means all programming languages can understand it.

* json module

The json module in Python helps Python send and receive data in JSON format by converting:

Python object ➝ JSON string

JSON string ➝ Python object

```python
## Data Serialization
import json
data={'name':'Krish','age':25}

json_str=json.dumps(data) #dumps stands for "dump string".
# dumps() is a function from the json module that converts a Python object into a JSON formatted string.
# In this case, it will convert the Python object stored in data (like a dictionary or list) into a JSON string format.
print(json_str)
print(type(json_str))

parsed_data=json.loads(json_str) # loads() is a function from the json module that converts a JSON string into a Python object.
# In this case, it takes JSON formatted data (string) and converts it into a Python dictionary, list, etc.
print(parsed_data)
print(type(parsed_data))
```

```python
## csv module

import csv
 
with open('example.csv',mode='w',newline='') as file:  # open() function file ko write mode ('w') me open karta hai. 
    # Agar file pehle se exist nahi karti, to ye nayi file create kar deta hai. 
    # # Agar file already exist karti hai, to purani file ka data delete (overwrite) ho jata hai. """
    writer=csv.writer(file) #csv.writer() creates a writing tool, and through that tool we can use writerow() to write rows into a CSV file.
    writer.writerow(['name','age'])
    writer.writerow(['Krish',32])

with open('example.csv',mode='r') as file: #open() function file ko read mode ('r') me open karta hai.Ye file se data padhne (read karne) ke liye use hota hai.
    reader=csv.reader(file) #reader() is a function of the csv module that creates a reader object to read data from a CSV file row by row.
    for row in reader:
        print(row)
```

```python
## datetime module 
from datetime import datetime,timedelta
# Inside the datetime module, datetime is a class

now=datetime.now() # now() is a function of the datetime class (from the datetime module) that returns the current date and time of the system. It gives the present local date and time as a datetime object.
print(now)

yesterday=now-timedelta(days=1) #timedelta is a class from the datetime module that represents a time difference.

print(yesterday)
```

```python
## time module
import time
print(time.time()) # prints current timestamp — total seconds (with decimal precision) since 1 January 1970.
time.sleep(2) # program waits for 2 seconds
print(time.time()) 
```

```python
## Regular expresiion
import re # This module is used to find the pattern in in text.

pattern=r'\d+' # r'\d+' is a regular expression pattern where \d matches any digit (0–9) and + means one or more times, so it matches a sequence of one or more digits.
text='There are 123 apples 456'
match=re.search(pattern,text)  # search() function text me pattern ko dhundta hai.
print(match.group())

## This code uses a regular expression to find the first number in the text and prints it.
```
