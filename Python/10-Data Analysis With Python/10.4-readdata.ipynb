### Reading Data From Different Sources


```python
import pandas as pd
from io import StringIO # Import StringIO to convert a string into a file-like object because pandas.read_json() normally reads JSON from a file.
Data = '{"employee_name": "James", "email": "james@gmail.com", "job_profile": [{"title1":"Team Lead", "title2":"Sr. Developer"}]}' # JSON data stored as a string
df=pd.read_json(StringIO(Data)) # read_json() reads JSON formatted data and converts it into a pandas DataFrame
# StringIO converts the JSON string into a file-like object so pandas can read it
```

```python
df
```

```python
df.to_json() # # Convert DataFrame to JSON format where each column contains index-value pairs.
```

```python
df.to_json(orient='index')
# Convert DataFrame to JSON where each row index contains its column as key-value pairs
```

```python
df.to_json(orient='records')
# Convert DataFrame to JSON as a list of row dictionaries (records format)
```

```python
df=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data",header=None)
# Read CSV dataset from a URL and convert it into a pandas DataFrame
# header=None is used because pandas by default treats the first row as column names.
# This dataset does not contain column headers, so we set header=None.
```

```python
df.head()
```

```python
df.to_csv("wine.csv") ## Save the DataFrame into a CSV file named 'wine.csv'
```

```python
url="https://www.fdic.gov/resources/resolutions/bank-failures/failed-bank-list/"

df=pd.read_html(url) # read_html() is a function of Pandas that reads all HTML tables from the webpage and store them as a list of DataFrames.
```

```python
df[0] ## Access the first table from the list . df[1] means 2nd table , df[3] means 3rd table....
```

```python
url="https://en.wikipedia.org/wiki/Mobile_country_code"
pd.read_html(url,match="Country",header=0)[0]
# Read tables from the given webpage URL
# 'match="Country"' tells pandas to select only the table that contains the word "Country"
# 'header=0' means the first row of the table will be used as column names
# read_html() returns a list of DataFrames, so [0] selects the first matching table 
```

```python
df_excel=pd.read_excel('data.xlsx') # read_excel() reads data from an Excel (.xlsx) file and converts it into a pandas DataFrame
```

```python
df_excel.to_pickle('df_excel')
# Save the DataFrame as a pickle file for fast storage and later use
# pandas to_pickle() method saves the DataFrame in pickle format.
# Pickle is a Python format that stores objects in a binary file.
# Advantages:
# 1. DataFrame loads very fast.
# 2. The complete Python object structure remains the same.
```

```python
pd.read_pickle('df_excel')
# read_pickle() loads a DataFrame stored in pickle format.
# It reads the binary pickle file and reconstructs the original Python object (DataFrame).
# This method is faster than reading CSV or Excel files.
```
