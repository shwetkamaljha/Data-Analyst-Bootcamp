```python
import seaborn as sns
```

```python
### Basic Plotting With Seaborn
tips=sns.load_dataset('tips') # Load the built-in 'tips' dataset from Seaborn.Seaborn comes with some built-in datasets for practice and examples tips.
tips
```

### Relationship between Seaborn and Matplotlib

Seaborn is based on Matplotlib.

Every Seaborn plot internally uses Matplotlib’s figure and axes.

This means the type of plot is the same in both, but the visual style is different — Seaborn looks nicer by default.

Why we still use Matplotlib with Seaborn

Seaborn makes plotting easier, but sometimes you need extra control, like:

Adding a title → plt.title("...")

Adding axis labels → plt.xlabel("..."), plt.ylabel("...")

Showing the plot → plt.show()

Changing figure size → plt.figure(figsize=(..,..))

These are all Matplotlib commands, so you still need to import matplotlib.pyplot (plt).

### Scatter Plot

A scatter plot is a type of plot used to show the relationship between two numerical variables.

Each data point is represented as a marker (dot, circle, etc.) on the graph.

The X-axis represents one variable and the Y-axis represents the other.

Lines are not drawn by default —  
## only the points are plotted. ##

Useful for identifying trends, correlations, clusters, or outliers in data.

```python
##create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

### Line Plot in Seaborn

A line plot shows the relationship between two numerical variables.

It connects data points with lines to show trends or changes over a continuous variable (like time).

In Seaborn, line plots are created using the sns.lineplot() function.

Works well with time series or ordered data.

```python
## Line Plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line Plot of Total bill by size")
plt.show()
```

### Categorical Plots

Categorical plots are used to visualize data where one variable is categorical (like names, types, days, etc.) and the other is numerical (like sales, scores, counts).

They help compare values across different categories.

### Bar Plot

Bar plot is the most common type of categorical plot.

It shows rectangular bars where the height (or length) of each bar represents the value for that category.

Bars can be vertical or horizontal.

```python
## Categorical Plots
## BAr Plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar Plot of Total Bill By Day')
plt.show()
```

### Box Plot (Whisker Plot) (a type of categorical plot used to visualize the distribution of numerical data)

Shows a summary of the data (quick view of data spread and key statistics):

Median (middle value of the data)

Quartiles (Q1, Q3) (data divided into four equal parts)

Whiskers (minimum and maximum values excluding outliers)

Outliers (data points that are unusually high or low)

```python
## Box Plot
sns.boxplot(x="day",y='total_bill',data=tips)
```

### Violin Plot (a categorical plot used to visualize the distribution of numerical data)

It combines a Box Plot and a KDE (Kernel Density Estimation) plot (shows both summary statistics and the full distribution shape).

Shows:

Median (middle value of the data)

Quartiles (Q1, Q3) (middle 50% of the data)

Density / Shape of distribution (how data is spread, peaks, and skewness)

Outliers (data points that are unusually high or low)

```python
## Violin Plot

sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
### Histograms
sns.histplot(tips['total_bill'],bins=10,kde=True)
# sns.histplot() → creates a histogram of numerical data.
# tips['total_bill'] → selects the column total_bill from the tips dataset.
# bins=10 → divides the data range into 10 equal intervals (bins).
# kde=True → overlays a Kernel Density Estimate (smooth curve) on top of the histogram, showing the distribution shape of the data.
```

### A KDE Plot is a plot that shows how the data values are spread or distributed.

It draws a smooth curve instead of bars (like a histogram).

The height of the curve shows where most of the data points are (high points = many values, low points = few values).

It can be used for one variable (univariate) or two variables (bivariate).

```python
## KDE Plot
sns.kdeplot(tips['total_bill'],fill=True , color='pink')
```

### sns.pairplot() → creates a matrix of plots showing relationships between all numerical variables in a dataset.

It automatically generates:

Scatter plots for every pair of numerical columns

Histograms or KDE plots on the diagonal (for single-variable distributions)

This helps you quickly see correlations, trends, and distributions for all variables at once.

```python
# Pairplot
sns.pairplot(tips)
```

### Heatmap

A Heatmap is a data visualization technique that represents numerical data using different colors in matrix to show the intensity, patterns, or relationships between variables.

In Data Analysis and Machine Learning, heatmaps are commonly used to visualize a correlation matrix, which helps to understand how strongly variables are related to each other.

### Correlation Matrix in Heatmap (Data Science)

A Correlation Matrix Heatmap is a visualization that shows the correlation values between multiple variables using colors.
It helps data scientists quickly understand the strength and direction of relationships between features in a dataset.

```python
## Heatmap
corr=tips[['total_bill','tip','size']].corr() # tips → a pandas DataFrame that contains restaurant data like bill amount, tip, number of people, etc.
# This selects only three columns from the dataset:
# total_bill → total bill amount of the table
# tip → tip given by the customer
# size → number of people sitting at the table

print(corr)

# .corr()
# This method calculates the correlation coefficient between the selected numerical columns.
# Correlation shows how strongly two variables are related.

# Correlation values range from -1 to +1:
# +1 → perfect positive relationship (both variables increase together)
# 0 → no relationship between variables
# -1 → perfect negative relationship (one increases while the other decreases)

# The result is stored in the variable 'corr'
# It becomes a correlation matrix (a table showing correlation between each pair of variables)


# Value	      Meaning
# 1	          perfect positive relation
# 0.7 – 0.9	  strong relation
# 0.4 – 0.7	  moderate relation
# 0.1 – 0.4	  weak relation
# 0	          no relation
# -1	          perfect negative relation
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')

# annot=True
# annot means annotation.
# It shows the actual numerical correlation values inside each cell of the heatmap.
# If annot=False, only colors will be shown without numbers.

# cmap means color map.
# It decides the color style of the heatmap.
# 'coolwarm' shows:
# blue colors → lower or negative correlation
# red colors → higher positive correlation

# The heatmap will visually show how strongly the variables (total_bill, tip, size) are related to each other using colors.
```

```python
import pandas as pd
sales_df=pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
## Plot total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x='Product',y="Sales",data=sales_df)
# Internal working:
# Seaborn groups the data by the 'Product' column.
# Then it calculates the average (mean) of the 'Sales' values for each product.
plt.title('Total Sales by Product')
plt.xlabel('Product Details')
plt.ylabel('Total Sales')
plt.show()
```

```python
## Plot total sales by Region
plt.figure(figsize=(10,6))
sns.barplot(x="Product" , y='Sales',data=sales_df,estimator=sum)
plt.title('Total Sales by Region')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()

# estimator=sum
# By default, seaborn barplot calculates the mean (average) of the Y values for each category.
# But here we changed it to 'sum'.
# So seaborn will calculate the total sales for each product.

# Internal working:
# Seaborn groups the dataset by 'Product'
# Then it adds (sums) the 'Sales' values for each product
# The height of each bar represents the total sales of that product.
```
