#### Pandas-DataFrame And Series
Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame. A Series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns).

```python
import pandas as pd 
```

```python
## Series
##A Pandas Series is a one-dimensional array-like object that can hold any data type with indexes. It is similar to a column in a table.

import pandas as pd
data=[1,2,3,4,5]
series=pd.Series(data) # pd.Series() converts list or array into a Pandas Series (1D labelled data structure with index and values)
print("Series \n",series)
print(type(series)) # Pandas Series objec
```

```python
## Create a Series from dictionary
data={'a':1,'b':2,'c':3}
series_dict=pd.Series(data)
print(series_dict)

# When creating a Series from dictionary: keys become index and values become data
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
# In pd.Series(), data and index length must be equal.
# Otherwise Pandas raises ValueError.
# Exception: when using dictionary, missing index values become NaN.
```

# # DataFrame is a 2D labelled data structure in pandas with rows and columns, similar to a table or Excel sheet.

```python
## Dataframe
## create a Dataframe from a dictionary of list

data={
    'Name':['Krish','John','Jack'],
    'Age':[25,30,45],
    'City':['Bangalore','New York','Florida'],
    'Domain':['DataAnalyst' ,'DataScientist' , 'ML Engineer']
}
df=pd.DataFrame(data)
print(df)
print(type(df))
# In pandas DataFrame, when using a dictionary of lists, all columns must have the same number of values.
# Otherwise pandas raises: ValueError: All arrays must be of the same length
```

```python
## Create a Data frame From a List of Dictionaries

data=[
    {'Name':'Krish','Age':32,'City':'Bangalore'},
    {'Name':'John','Age':34,'City':'Bangalore'},
    {'Name':'Bappy','Age':32,'City':'Bangalore' , 'Domain':'DataAnalytics'},
    {'Name':'JAck','Age':32}
    
]
df=pd.DataFrame(data)
print(df)
print(type(df))
## pd.DataFrame(list_of_dict) creates a DataFrame where each dictionary represents one row and dictionary keys become column names.
# Missing keys are filled with NaN
```

```python
import pandas as pd

# Create a DataFrame with custom alphabetic index labels
data = {
    'Name': ['Krish', 'John', 'Ram'],
    'Age': [25, 30, 28],
    'City': ['Bangalore', 'New York', 'Delhi']
}

df = pd.DataFrame(data, index=['a', 'b', 'c'])  # Set alphabetic labels as index
print(df)
```

```python
# loc → label-based indexing (selects data using the index label)

print(df.loc['a'] ,"\n")        # Select the row whose index label is 'a'
print(df.loc['b'],"\n")        # Select the row whose index label is 'b'
print(df.loc['a':'b'],"\n")    # Select rows from label 'a' to 'b' (both included)
print(df.loc['a','Name']) # Select value at row label 'a' and column 'Name'
```

```python

# iloc → integer location indexing (selects data using row/column position)

print(df.iloc[0],"\n")         # Select the first row using position 0
print(df.iloc[1],"\n")         # Select the second row using position 1
print(df.iloc[0:2],"\n")       # Select rows from position 0 to 1 (2 is excluded)
print(df.iloc[0,1])       # Select value at first row and second column
```
