Intro to Tuple
Tuples are ordered collections of items that are immutable. They are similar to lists but their immutability makes them diiferent.

```python
## Creating a Tuple 
emp_tup = ()
print(emp_tup)
print(type(emp_tup))
```

```python
number = tuple([1, 2, 3 , 3, 5 , 6 , 3 , 7 , 9]) ## Convert a list to a tuple
print(number)
print(type(number))
```

```python
mix_tup=tuple([1, 2, "Dog" , 4 , "Cat" , 9 , "Horse" ]) ## Convert a tuple to a list
print(mix_tup)
```

```python
# Accessing Tuple Elements
print(number[0]) # Output: 1
print(number[1]) # Output: 2
print(number[::-1]) # Output: (9, 7, 6, 5, 3, 2, 1)
print(number[1:5]) # Output: (2, 3, 5, 6)
print(number[:4]) # Output: (1, 2, 3, 5)
```

```python
## Tuple Operations
concatenation_tup = number + mix_tup
print(concatenation_tup)

```

```python
repetition_tup = mix_tup * 2
print(repetition_tup)
```

```python
number * 3
```

```python
##Tuple Methods
print(number)
print(number.count(3)) # Output: 3
print(number.index(5)) # Output: 4
```

```python
# packing and unpacking a tuple
## Packing a tuple
my_tuple = 1, "Hello", 3.14
print(my_tuple)
## Unpacking a tuple
a , b , c = my_tuple 
print(a) # Output: 1
print(b) # Output: Hello
print(c) # Output: 3.14
```

```python
## Unpacking with * operator
numbers = (1, 2, 3, 4, 5)
first, *middle, last = numbers
print(first) # Output: 1
print(middle) # Output: [2, 3, 4]       
print(last) # Output: 5
```

```python
# Nested Tuples
nestd_tup = tuple([ [1,2,3 ] , [6 , 7 , 8 , 9] , [1 , "Hello " , 3.14 , "c"]])
print(nestd_tup)
print(nestd_tup[0]) # Output: [1, 2, 3]
print(nestd_tup[1][2]) # Output: 8
print(nestd_tup[0][2]) # Output: 3
print(nestd_tup[2][0:3]) # Output: [1 ,'Hello ', 3.14]

```

```python
##Iterating through a tuple
for sub_tup in nestd_tup:
    for item in sub_tup:
        print(item , end=" ")
    print()
```

* Conclusion 

Tuples are versatile and useful in many real-world scenarios where an immutable and ordered collection of item is required . They are commonly used in data structures , function arguments and return values and as dictionary keys  . Understanding how to leverage tuples effectively can improve the efficiency and readability of your Python Code.
