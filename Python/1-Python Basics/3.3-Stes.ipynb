```python
my_set1 = set([1, 2, 3, 4, 5, 1, 2, 3])
print(my_set1) # Output: {1, 2, 3, 4, 5}
```

```python
## Basic Set Operations
## Adding and Removing Elements
my_set1.add(6)
print(my_set1) # Output: {1, 2, 3, 4, 5, 6} 
my_set1.add(6) # Adding duplicate element
print(my_set1) # Output: {1, 2, 3, 4, 5, 6} (No change)
my_set1.remove(3)
print(my_set1) # Output: {1, 2, 4, 5, 6}
my_set1.discard(10) # Discarding non-existent element (no error)
print(my_set1) # Output: {1, 2, 4, 5
```

```python
remvd_ele = my_set1.pop() # Remove and return a random element
print(f"Removed element: {remvd_ele}") # Output: Removed element: 1
print(my_set1) # Output: {2, 4, 5, 6} (One element removed)
my_set1.clear() # Remove all elements from the set
print(my_set1) # Output: set() (Empty set)
```

* Pop()

The pop() function removes any one random element from the set because a set is unordered, so there is no fixed element that will be removed.

```python
## Set Membership test
my_set = { 1 , 2, 3, 4 , 5}
print(3 in my_set) # Output: True
print(10 in my_set) # Output: False
```

```python
## Mathematical Set Operations
A = {1, 2, 3, 4, 5} 
B = {4, 5, 6, 7, 8}
print(A.union(B)) # Output: {1, 2, 3, 4, 5, 6, 7, 8}
print(A.intersection(B)) # Output: {4, 5}
print(A.difference(B)) # Output: {1, 2, 3}
print(A.symmetric_difference(B)) # Output: {1, 2, 3, 6, 7, 8}
A.intersection_update(B)
print(A) # Output: {4, 5} (A is now updated to the intersection of A and B)
```

```python
S1 = {1 , 2 , 3 , 4  , 5 , 6 , 7}
S2 = {3 , 4 , 5 , 6 , 7 }
S3 = {8 , 9 , 10}
print(S2.issubset(S1)) # Output: True (S2 is a subset of S1)
print(S1.issuperset(S2)) # Output: True (S1 is a superset of S2)
print(S1.isdisjoint(S3)) # Output: True (S1 and S3 have no common elements)
print(S1.issubset(S3)) # Output: False (S1 is not a subset of S3)
```

* split()

 Split method is used to split the text into individual words based on whitespace and convert it into a list


```python
#Convert unique word in Text
text =" Data analysis is the process of inspecting, cleansing, transforming, and modeling data to discover useful information, inform conclusions, and support decision-making."
words=text.split() # Split method is used to split the text into individual words based on whitespace and convert it into a list
print(words) # Output: ['Data', 'analysis', 'is', 'the', 'process', 'of', 'inspecting,', 'cleansing,', 'transforming,', 'and', 'modeling', 'data', 'to', 'discover', 'useful', 'information,', 'inform', 'conclusions,', 'support', 'decision-making.']
unique_words = set(words) # Convert list of words to a set to get unique words
print(unique_words) # Output: {'Data', 'analysis', 'is', 'the', 'process', 'of', 'inspecting,', 'cleansing,', 'transforming,', 'and', 'modeling', 'data', 'to', 'discover', 'useful', 'information,', 'inform', 'conclusions,', 'support', 'decision-making.'}
print(f"Number of unique words: {len(unique_words)}") # Output: Number of unique words: 20
```
