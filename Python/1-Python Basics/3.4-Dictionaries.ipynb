Intro to Dictionaries

Dictionaries are mutable and  unordered collections of items . They store data in key values pairs . Keys must be unique and immutable (e.g. strings , numbers ,or tuples) while values can be of any type.

```python
# These are the two methods to create a dictionary.
d={}
print(type(d))
emp_d = dict()
print(type(emp_d))
```

```python
# Acessing Dictionary Elements
std ={ "name": "John", "age": 25, "city": "New York"}
print(std) # Output: {'name ': 'John', 'age': 25, 'city': 'New York'}
print(std["name"]) # Output: John
print(std["age"]) # Output: 25
print(std["city"]) # Output: New York   
```

```python
#Accessinh using get() method
print(std.get("name")) # Output: John
print(std.get("age")) # Output: 25
print(std.get("country", "Nhi mila")) # Output: Nhi mila (since 'country' key does not exist, it returns the default value "Nhi mila")
print(std.get("Grade")) # Output: None (since 'Grade' key does not exist and no default value is provided, it returns None) 
```

```python
# Duplicate keys in a dictionary will overwrite the previous value. In this example, the second 'name' key with the value 'Bob' overwrites the first 'name' key with the value 'Alice'.
std1 = {"name": "Alice", "age": 30, "name": "Bob"}
print(std1) # Output: {'name': 'Bob', 'age': 30} (the second 'name' key overwrites the first one)
```

```python
# Modifying Dictionary
print(std) # Output: {'name ': 'John', 'age': 25, 'city': 'New York'}
std["age"] = 26 # Update age
print(std) # Output: {'name ': 'John', 'age': 26, 'city': 'New York'}
std["country"] = "USA" # Add new key-value pair 
print(std) # Output: {'name ': 'John', 'age': 26, ' city': 'New York', 'country': 'USA'}
```

```python
# Deleting Dictionary Elements
del std["age"] # Delete the 'age' key-value pair
print(std) # Output: {'name ': 'John', 'city': 'New York',
```

```python
#Dictionary Methods
print(std.keys()) # print the keys of the dictionary
print(std.values()) # print the values of the dictionary
print(std.items()) # print the key-value pairs of the dictionary as tuples in a list
```

```python
custmr = {"name": "Alice", "age": 30, "city": "New York"}
```

```python
## Shallow Copy 
custmr_copy = custmr
print(custmr)
print(custmr_copy) # Output: {'name': 'Alice', 'age': 30, 'city': 'New York'}
```

```python
custmr["name"] = "Bob"
print(custmr) # Output: {'name': 'Bob', 'age': 30, 'city': 'New York'}
print(custmr_copy) # Output: {'name': 'Bob', 'age': 30, 'city': 'New York'} (since custmr_copy is a shadow copy of custmr, changes to custmr also affect custmr_copy)
```

```python
custmr_copy1 = custmr.copy() # This creates a shallow copy of the dictionary
print(custmr_copy1) # Output: {'name': 'Bob', 'age': 30, 'city': 'New York'}    
print(custmr) # Output: {'name': 'Bob', 'age': 30, 'city': 'New York'}

```

```python
custmr["name"] = "Bob_1"
print(custmr_copy1) # Output: {'name': 'Bob', 'age': 30, 'city': 'New York'} (since custmr_copy1 is a shallow copy of custmr, changes to custmr do not affect custmr_copy1)
print(custmr) # Output: {'name': 'Bob_1', 'age': 30, 'city': 'New York'}
```

```python
## Iterating through a dictionary 
## You can use loop to iterate over dictionary keys, values and Items.

## Iterating over keys 
for key in custmr.keys():
    print(key)
```

```python
# Iterating over values
for value in custmr.values():
    print(value)
```

```python
# Iterating over Items of a dictionary
for item in custmr.items():
    print(item)
```

```python
for key , value in custmr.items():
    print(f"{key} : {value}") # This is a print statement that uses an f-string to format the output. It will print each key and its corresponding value in the dictionary in a readable format.
```

```python
# Nested Dictionary
students = {
    "student1": {"name": "Alice", "age": 20, "city": "New York"},
    "student2": {"name": "Bob", "age": 22, "city": "Delhi"}
}
print(students) # Output: {'student1': {'name': 'Alice', 'age': 20, 'city': 'New York'}, 'student2': {'name': 'Bob', 'age': 22, 'city': 'Los Angeles'}}
```

```python
## Accessing nested dictionary elements
print(students["student1"]) # Output: {'name': 'Alice', 'age': 20, 'city': 'New York'}
print(students["student1"]["name"]) # Output: Alice
print(students["student2"]["city"]) # Output: Delhi
```

```python
## Iterating over nested dictionary
for student_id, details in students.items():
    print(f"{student_id}: {details}")
    for key, value in details.items():
        print(f"  {key}: {value}")
```

```python
## Dictionary Comprehension
squares = {x: x**2 for x in range(1, 6)}
print(squares) # Output: {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
```

```python
## Conditional Dictionary Comprehension
even_squares = {x: x**2 for x in range(1, 11) if x % 2 == 0}
print(even_squares) # Output: {2: 4, 4: 16, 6: 36, 8: 64, 10: 100} (this creates a dictionary of squares for even numbers from 1 to 10)
```

```python
## Practical Examples
## Use a dictionary to count the frequency of elements in a list
numbers = [1, 2, 2, 3, 3, 3, 4]
freq = {}
for number in numbers:
    if number in freq:
        freq[number] += 1 #freq[number] = freq[number] + 1
    else:
        freq[number] = 1
print(freq) # Output: {1: 1, 2: 2, 3: 3, 4: 1} (this creates a dictionary where the keys are the unique numbers from the list and the values are their corresponding frequencies)
```

```python
## Merge two dictionaries
dict1 = {"a": 1, "b": 2}
dict2 = {"b": 3, "c": 4}
merged_dict = {**dict1, **dict2} # This merges dict1 and dict2 into a new dictionary. If there are duplicate keys, the value from dict2 will overwrite the value from dict1. In this case, the key 'b' has a value of 2 in dict1 and a value of 3 in dict2, so the merged dictionary will have 'b': 3.
print(merged_dict) # Output: {'a': 1, 'b': 3 ,  'c': 4} (the merged dictionary contains all key-value pairs from both dictionaries, with the value of 'b' being 3 from dict2)
```
