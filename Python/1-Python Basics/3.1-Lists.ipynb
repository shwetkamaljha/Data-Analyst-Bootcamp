Basic List Comprehension

List Intro 
* Lists are ordered , mutable collections of items.
* They can contain items of different data types.

```python
### Acesing List Elements
lst = [ "apple ", "banana ", "cherry " , "Guava" , "Mango" ]
print(lst[0]) # Output: apple
print(lst[1]) # Output: banana
print(type(lst))
print(lst[-1]) # Output: Mango
```

```python
## Slicing Lists
print(lst[1:4]) # Output: ['banana ', 'cherry ', 'Guava']
print(lst[:3]) # Output: ['apple ', 'banana ', 'cherry ']
print(lst[::-1]) # Output: ['Mango', 'Guava', 'cherry ', 'banana ', 'apple ']
print(lst[::2]) # Output: ['apple ', 'cherry ', 'Mango']
print(lst[::]) # Output: ['apple ', 'banana ', 'cherry ', 'Guava', 'Mango']
```

```python
### Modifying the list elements
print(lst)
lst[1] = "Strawberry"
print(lst)
lst[1:3] = ["Watermelon", "Pineapple"]
print(lst)
```

```python
# ### List Methods
lst.append("Orange")
print(lst)
lst.insert(2, "Kiwi") ## Inserts "Kiwi" at index 2
print(lst)
pop_ele=lst.pop()
print(pop_ele)
print(lst)
print(lst.index("Pineapple")) ## Returns the index of "Pineapple" in the list
lst.sort()
print(lst)
lst.clear() ## Clears the list

```

```python
lst1 = [ 1, 2, 3  , 2 , 4 , 5 , 2 , "skj" , "dfg" ]
print(lst1.count(2)) ## Counts the number of occurrences of 2 in the list
lst1.remove("skj") ## Removes the first occurrence of "skj" from the list
print(lst1)
lst1.reverse() ## Reverses the list
print(lst1)
```

```python
#List Comprehension
Lst =[]
for x in range (10):
    Lst.append(x**2)

print(Lst)

Lst1 = [ x**2 for x in range(10) ]
print(Lst1)
```

List Comprehension

* Basic Syntax             [expression for item in iterable]


* With conditional logic   [expression for item in iterable if condition]

* Nested List Comprehension [expression for item1 in iterable1 for item2 in iterable2]


```python
###Basic List Comprehension
square = [ num**2 for num in range(1, 11) ]
print(square)
```

```python
## List comprehension with condition
even_square = [ num**2 for num in range(1, 11) if num % 2 == 0]
print(even_square)
```

Nested List Comprehension

```python
lst=[1,2,3,4]
lst1=['a','b','c','d']
pair=[[i,j] for i in lst for j in lst1]
print(pair)
```

```python
## List comprehension with functin calls
def square(x):
    return x**2
squared_list = [ square(num) for num in range(1, 11) ]
print(squared_list)
```

```python
words = [" hello", "world", "python", "programming"]
lengths = [ len(word) for word in words ]
print(lengths) # Output: [6, 5, 6, 11]
```

# Conclusion 
List comprehensions are a powerful and concise way to create lists in Python. They are syntactically compact and can replace more verbose looping constructs. Understanding the syntax of list comprehension will help you write cleaner and more efficient Python Code.
